```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def u(n):
    return (n >= 0).astype(int)

def h(n, N):
    return np.sin(np.pi * n / 2) * (u(n) - u(n - 2 * N))

def H(w):
    e = np.e
    return (e ** (-1j * w) - e ** (-5j * w)) / (1 + e ** (-2j * w))
```

```python
def frequencyResponse(w):
    return np.absolute(H(w))

def phaseResponse(w):
    return np.angle(H(w))
```

```python
xLine = np.linspace(-10, 10, 10000)
yLine = frequencyResponse(xLine)
#yLine = np.ma.masked_where(yLine > 3, yLine)
#yLine = np.ma.masked_where(yLine < -3, yLine)
plt.plot(xLine, yLine)
```

```python
xLine = np.linspace(-10, 10, 50000)
yLine = phaseResponse(xLine)
yLine[:-1][np.diff(yLine) > 0] = np.nan
plt.plot(xLine, yLine)
```
